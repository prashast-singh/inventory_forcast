##### **b)** 
As in **A1**, use the `iris` dataset again. Create histograms of the features *'petal length'*, *'sepal length'*, and *'petal width'*. Experiment with the bin width and choose one that works well for you. Then, create two scatter plots: *'petal length' vs. 'petal width'* and *'petal length' vs. 'sepal length'*. Make sure your plots are displayed clearly and do not overlap (use `plt.subplots(x, y, figsize=(p, q))` and `plt.tight_layout()`). Use <a href="https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.pearsonr.html">`pearsonr()`</a> from `scipy.stats` to calculate the Pearson correlation for both column pairs. Interpret your results. Do the scatter plots support the computed values? What structure in the data can be inferred from the histograms? Could you enrich the scatter plots with additional information that helps explain the patterns? **(3 pts)**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from scipy.stats import pearsonr


iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=[name.replace(' (cm)', '').replace(' ', '_') for name in iris.feature_names])
iris_df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)


features = ['petal_length', 'sepal_length', 'petal_width']
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

for ax, feature in zip(axes, features):
    ax.hist(iris_df[feature], bins=15, color='skyblue', edgecolor='black')
    ax.set_title(f'Histogram of {feature.replace("_", " ").title()}')
    ax.set_xlabel(feature.replace('_', ' ').title())
    ax.set_ylabel('Frequency')

plt.tight_layout()
plt.show()


fig, axes = plt.subplots(1, 2, figsize=(12, 5))


sns.scatterplot(data=iris_df, x='petal_length', y='petal_width', hue='species', ax=axes[0])
axes[0].set_title('Petal Length vs Petal Width')
axes[0].set_xlabel('Petal Length (cm)')
axes[0].set_ylabel('Petal Width (cm)')

sns.scatterplot(data=iris_df, x='petal_length', y='sepal_length', hue='species', ax=axes[1])
axes[1].set_title('Petal Length vs Sepal Length')
axes[1].set_xlabel('Petal Length (cm)')
axes[1].set_ylabel('Sepal Length (cm)')

plt.tight_layout()
plt.show()


corr_pw, pval_pw = pearsonr(iris_df['petal_length'], iris_df['petal_width'])
corr_sl, pval_sl = pearsonr(iris_df['petal_length'], iris_df['sepal_length'])


print("Pearson Correlation Results:")
print(f"Petal Length vs Petal Width: r = {corr_pw:.3f}, p-value = {pval_pw:.3e}")
print(f"Petal Length vs Sepal Length: r = {corr_sl:.3f}, p-value = {pval_sl:.3e}")
```

**Note your interpretations here:**

  [Petal Length and Petal Width histograms show distinct peaks, suggesting clear clustering — particularly one tight group and two more spread out.
  
This strongly implies the presence of separable species groups, as we know the dataset contains three species (setosa, versicolor, virginica).

Sepal Length shows a more continuous and overlapping distribution, indicating it’s less discriminative between species than petal measurements.

The strong r values indicate a linear relationship between petal length and both petal width and sepal length.
Very low p-values (< 0.05) indicate these correlations are statistically significant.

]

##### **c)** 
Create a scatter plot of column `F1` against column `F2` from the file `data.csv` and again compute the Pearson correlation coefficient along with the $p$-value using the method <a href="https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.pearsonr.html">`pearsonr()`</a>. Interpret your result: use the **scatter plot** to **critically assess** the Pearson correlation value. State the null hypothesis $H_0$ of the statistical test for which the $p$-value is returned by the `scipy` method, and make a decision based on it. Then determine what would need to be checked to ensure the validity of that result (you do not need to perform this check in this task). **(3 pts)**

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import pearsonr


df = pd.read_csv('data.csv')

# Scatter plot of F1 vs F2
plt.figure(figsize=(8, 6))
plt.scatter(df['F1'], df['F2'], alpha=0.7)
plt.title('Scatter Plot of F1 vs F2')
plt.xlabel('F1')
plt.ylabel('F2')
plt.grid(True)
plt.show()


corr_coef, p_value = pearsonr(df['F1'], df['F2'])

print(f"Pearson correlation coefficient: {corr_coef:.4f}")
print(f"P-value: {p_value:.4e}")


if p_value < 0.05:
    print("The p-value is less than 0.05, so we reject the null hypothesis.")
    print("There is statistically significant evidence of a linear correlation between F1 and F2.")
else:
    print("The p-value is greater than or equal to 0.05, so we fail to reject the null hypothesis.")
    print("There is no statistically significant evidence of a linear correlation between F1 and F2.")


print("\nTo ensure validity of the Pearson correlation test result, the following assumptions should be checked:")
print("- Linearity: The relationship between F1 and F2 should be linear (can be visually assessed from the scatter plot).")
print("- Normality: Both F1 and F2 should be approximately normally distributed.")
print("- Homoscedasticity: The variability of F2 should be consistent across the values of F1.")
print("- No significant outliers: Outliers can distort the correlation.")


print("\nNull Hypothesis (H0): There is no linear correlation between F1 and F2 (correlation coefficient = 0).")
```

**Note your interpretations here:**

The correlation coefficient of 0.9332 indicates a very strong positive linear relationship between F1 and F2.
As F1 increases, F2 also increases (or becomes less negative), which is visually supported by the upward trend in the scatter plot.

The p-value is effectively 0, far below the typical threshold of 0.05.
Decision: Reject the null hypothesis
