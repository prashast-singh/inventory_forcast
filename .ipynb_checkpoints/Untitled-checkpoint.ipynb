```python
# Cell 1: Imports & Config
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from sklearn.metrics import mean_absolute_percentage_error

# Jupyter magic
%matplotlib inline
plt.style.use('default')



```

```python
# Load data
df = pd.read_csv("sales_data 2.csv", parse_dates=['Date'])
# Derive Month period (end of month timestamp)
df['Month'] = df['Date'].dt.to_period('M').dt.to_timestamp('M')

print("Date range:", df['Month'].min(), "to", df['Month'].max())
print("Total rows:", len(df))
```

```python
# Cell 3: Build Leaf-Level Series
# One timeseries per (Category, Subcategory, Country, State)
leaf = (
    df
    .groupby(['Month','Product_Category','Sub_Category','Country','State'])['Revenue']
    .sum()
    .unstack(['Product_Category','Sub_Category','Country','State'])
)

# Reindex to all month-ends (2011-01-31 → 2016-07-31)
all_months = pd.date_range(leaf.index.min(), leaf.index.max(), freq='M')
leaf = leaf.reindex(all_months)

# Name MultiIndex levels
leaf.columns.set_names(
    ['Product_Category','Sub_Category','Country','State'],
    inplace=True
)

leaf.head()
```

```python
# Cell 4: Impute the 2014 Aug–Dec Gap with Seasonal Averages
# Force true NaNs for that known gap
leaf.loc['2014-08-31':'2014-12-31', :] = np.nan

def seasonal_fill(series):
    month_mean = series.groupby(series.index.month).transform('mean')
    return series.fillna(month_mean)

leaf_imputed = leaf.apply(seasonal_fill)

# Plot a sample series to verify
plt.figure(figsize=(8,3))
leaf_imputed[('Bikes','Mountain Bikes','United States','California')].plot()
plt.title('Imputed – Bikes | Mountain Bikes | US:CA')
plt.ylabel('Revenue')
plt.show()
```

```python
# Cell 5: Forecast Each Leaf (12-month horizon) with Fallback
h = 12
forecast_dict = {}

for col in leaf_imputed.columns:
    ts = leaf_imputed[col].dropna()
    
    # If ≥24 months of data → ETS
    if len(ts) >= 24:
        model = ExponentialSmoothing(
            ts, trend='add', seasonal='add', seasonal_periods=12
        )
        fit = model.fit(optimized=True)
        fc = fit.forecast(h)
        
    else:
        # Seasonal-naive: repeat last 12 months (pad with mean if <12)
        last_12 = ts[-12:].copy()
        if len(last_12) < 12:
            pad_len = 12 - len(last_12)
            pad_idx = [ts.index[-1] + pd.offsets.MonthEnd(i) for i in range(1, pad_len+1)]
            pad_vals = np.repeat(ts.mean(), pad_len)
            last_12 = pd.concat([last_12, pd.Series(pad_vals, index=pad_idx)])
        
        idx_fc = pd.date_range(ts.index[-1] + pd.offsets.MonthEnd(1), periods=h, freq='M')
        vals = np.tile(last_12.values, int(np.ceil(h/12)))[:h]
        fc = pd.Series(vals, index=idx_fc)
    
    forecast_dict[col] = fc

# Build forecast DataFrame
fc = pd.DataFrame(forecast_dict)
fc.index.name = 'Month'
```

```python
# Cell 6: Aggregate & Plot History vs. Forecast (Global & Bikes)
# Aggregate
hist_total   = leaf_imputed.sum(axis=1)
hist_bikes   = leaf_imputed['Bikes'].sum(axis=1)
fc_total     = fc.sum(axis=1)
fc_bikes     = fc.groupby(level=0, axis=1).sum()['Bikes']

# Ensure proper datetime indices
for s in [hist_total, hist_bikes, fc_total, fc_bikes]:
    s.index = pd.to_datetime(s.index)

# Plot
fig, ax = plt.subplots(1,1, figsize=(10,5))

ax.plot(hist_total.index, hist_total.values,      label='History – Total')
ax.plot(fc_total.index,   fc_total.values,   '--', label='Forecast – Total (12m)')
ax.plot(hist_bikes.index, hist_bikes.values, ':',  label='History – Bikes')
ax.plot(fc_bikes.index,   fc_bikes.values,   '-.', label='Forecast – Bikes (12m)')

# Format x-axis as dates
ax.xaxis.set_major_locator(mdates.AutoDateLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
fig.autofmt_xdate()

ax.set_title('Total & Bikes Revenue: History vs. 12-Month Forecast')
ax.set_xlabel('Month')
ax.set_ylabel('Revenue')
ax.legend()
plt.tight_layout()
plt.show()
```

```python
# Cell 7: Example Rolling-Origin Backtest on One Leaf (fixed start handling)
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from sklearn.metrics import mean_absolute_percentage_error

def backtest(series, start, h):
    # ensure datetime
    start_ts = pd.to_datetime(start)
    # training up to and including start_ts
    train = series[:start_ts]
    # test window: next h months
    test_start = start_ts + pd.offsets.MonthEnd(1)
    test_end   = start_ts + pd.offsets.MonthEnd(h)
    test       = series[test_start:test_end]
    
    # fit model on train
    fit = ExponentialSmoothing(
        train,
        trend='add',
        seasonal='add',
        seasonal_periods=12
    ).fit(optimized=True)
    
    # forecast h steps
    pred = fit.forecast(h)
    # compute MAPE only on overlapping dates
    common_idx = test.index.intersection(pred.index)
    mape = mean_absolute_percentage_error(test.loc[common_idx], pred.loc[common_idx])
    
    return mape, train, test, pred

# pick a sample leaf
col = ('Bikes','Mountain Bikes','United States','California')

mape, train, test, pred = backtest(
    leaf_imputed[col],
    start='2015-01-31',
    h=12
)

print(f"Backtest MAPE for {col}: {mape:.2%}")

plt.figure(figsize=(8,3))
plt.plot(train.index, train.values,      label='Train')
plt.plot(test.index,  test.values,       label='Test')
plt.plot(pred.index,  pred.values, '--', label='Forecast')
plt.title(f'Backtest – {col}')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Cell 8: Revenue → Units → Inventory Rules (fixed Series naming)

# a) Compute historical avg price per month and name it
avg_price = (
    df.groupby('Month')['Revenue'].sum()
      .div(df.groupby('Month')['Order_Quantity'].sum())
      .rename('Avg_Price')
)

# b) Build DataFrame with your total revenue forecast (fc_total) and join avg_price
unit_fc = pd.DataFrame({'Revenue_Fc': fc_total}).join(avg_price, how='left')

# c) Estimate forecasted units demand
unit_fc['Units_Fc'] = unit_fc['Revenue_Fc'] / unit_fc['Avg_Price']

# d) Historical demand stats (units per month)
hist_units = df.groupby('Month')['Order_Quantity'].sum()
mu, sigma = hist_units.mean(), hist_units.std()

# e) Safety Stock & Reorder Point (2-month lead time, 95% service level)
LT, z = 2, 1.65
SS  = z * sigma * np.sqrt(LT)
ROP = mu * LT + SS

print(f"Avg monthly units: {mu:.0f}")
print(f"Safety stock:      {SS:.0f} units")
print(f"Reorder point:     {ROP:.0f} units")

# f) Suggest order quantities (units beyond ROP)
suggest = unit_fc[['Units_Fc']].copy()
suggest['Reorder_Qty'] = (suggest['Units_Fc'] - ROP).clip(lower=0)

suggest.head(12)
```

```python
import pandas as pd
import numpy as np
from statsmodels.tsa.holtwinters import ExponentialSmoothing

import warnings

warnings.filterwarnings('ignore')

# --- Configuration ---
DATA_PATH = 'sales_data 2.csv'
HORIZON = 6  # months ahead
INITIAL_TRAIN_MONTHS = 24
ALPHA_CROSTON = 0.1
SEGMENT_THRESHOLDS = {'A': 0.8, 'B': 0.95, 'C': 1.0}

# --- Data Loading and Preprocessing ---
def load_and_prepare(path):
    df = pd.read_csv(path, parse_dates=['Date'])
    df.set_index('Date', inplace=True)
    # Aggregate to monthly revenue per SKU
    monthly = (
        df.groupby([pd.Grouper(freq='M'), 'Product'])['Revenue']
          .sum()
          .reset_index(name='Monthly_Revenue')
    )
    ts = monthly.pivot(index='Date', columns='Product', values='Monthly_Revenue').fillna(0)
    return ts

# --- ABC Segmentation ---
def abc_segmentation(ts):
    total_rev = ts.sum()
    cum_share = total_rev.sort_values(ascending=False).cumsum() / total_rev.sum()
    segments = {}
    for product, share in cum_share.items():
        if share <= SEGMENT_THRESHOLDS['A']:
            segments[product] = 'A'
        elif share <= SEGMENT_THRESHOLDS['B']:
            segments[product] = 'B'
        else:
            segments[product] = 'C'
    return segments

# --- Forecasting Methods ---
def fit_ets(train, horizon):
    train = train.asfreq('M')
    model = ExponentialSmoothing(train,
                                 trend='add',
                                 damped_trend=True,
                                 seasonal='add',
                                 seasonal_periods=12)
    fit = model.fit(optimized=True)
    forecast = fit.forecast(horizon)
    return fit.fittedvalues, forecast

def croston_forecast(series, alpha, horizon):
    demand = series.values
    intervals = (demand > 0).astype(int)
    q = demand[demand > 0].mean() if any(demand>0) else 0
    p = intervals.mean() if len(intervals)>0 else 0
    for t in range(1, len(demand)):
        if demand[t] > 0:
            q = alpha * demand[t] + (1 - alpha) * q
            p = alpha * intervals[t] + (1 - alpha) * p
    f_val = q / p if p > 0 else 0
    idx = [series.index[-1] + pd.DateOffset(months=i) for i in range(1, horizon+1)]
    return pd.Series([f_val]*horizon, index=idx)

def seasonal_naive_forecast(train, horizon):
    train = train.asfreq('M')
    last_year = train[-12:]
    reps = (horizon // 12) + 1
    vals = np.tile(last_year.values, reps)[:horizon]
    idx = [train.index[-1] + pd.DateOffset(months=i) for i in range(1, horizon+1)]
    return pd.Series(vals, index=idx)

# --- Rolling Cross-Validation ---
def safe_mape(y_true, y_pred):
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    mask = y_true != 0
    if not mask.any():
        return np.nan
    return np.mean(np.abs((y_true[mask] - y_pred[mask]) / y_true[mask])) * 100


def rolling_cv(series, model_func, seg, initial_train=INITIAL_TRAIN_MONTHS,
               horizon=HORIZON, alpha=ALPHA_CROSTON):
    n = len(series)
    folds = (n - initial_train) // horizon
    errors = []
    for i in range(folds):
        end = initial_train + i * horizon
        train = series[:end]
        test = series[end:end+horizon]
        if seg == 'A':
            _, forecast = fit_ets(train, horizon)
        elif seg == 'B':
            forecast = croston_forecast(train, alpha, horizon)
        else:
            forecast = seasonal_naive_forecast(train, horizon)
        test_vals = test.values[:len(forecast)]
        pred_vals = forecast.values[:len(test)]
        mape = safe_mape(test_vals, pred_vals)
        if not np.isnan(mape):
            errors.append(mape)
    return np.nanmean(errors)

# --- Model Selection for B-segment ---
def compare_methods_b(ts, segments, sample_n=5):
    results = []
    b_skus = [p for p, s in segments.items() if s=='B'][:sample_n]
    for sku in b_skus:
        series = ts[sku]
        m_croston = rolling_cv(series, croston_forecast, 'B')
        m_seasonal = rolling_cv(series, seasonal_naive_forecast, 'C')
        pref = 'Croston' if (not np.isnan(m_croston) and m_croston <= m_seasonal) else 'SeasonalNaive'
        results.append({
            'Product': sku,
            'Croston_MAPE': round(m_croston,2),
            'SeasonalNaive_MAPE': round(m_seasonal,2),
            'Preferred': pref
        })
    return pd.DataFrame(results)

# --- Final Forecasting Pipeline ---
def generate_forecasts(ts, segments, method_map=None, horizon=HORIZON):
    """
    Generate forecasts for each product based on its ABC segment and optional method override for B-segment.
    Returns a DataFrame where columns are products and rows are forecasted timestamps.
    """
    forecasts = {}
    for product, seg in segments.items():
        series = ts[product]
        # ETS for A-segment
        if seg == 'A':
            _, fc = fit_ets(series, horizon)
        # B-segment: use override map to choose between Croston or seasonal naïve
        elif seg == 'B':
            pref = method_map.get(product) if method_map else None
            if pref == 'SeasonalNaive':
                fc = seasonal_naive_forecast(series, horizon)
            else:
                fc = croston_forecast(series, ALPHA_CROSTON, horizon)
        # C-segment: seasonal naïve
        else:
            fc = seasonal_naive_forecast(series, horizon)
        forecasts[product] = fc
    return pd.DataFrame(forecasts)

# --- Main Execution ---
if __name__ == '__main__':
    ts = load_and_prepare(DATA_PATH)
    segments = abc_segmentation(ts)

    # Evaluate B-segment methods and build override map
    comp_df = compare_methods_b(ts, segments)
    comp_df.to_csv('b_segment_method_comparison.csv', index=False)
    method_map = comp_df.set_index('Product')['Preferred'].to_dict()

    # Generate final forecasts for next horizon months
    final_fc = generate_forecasts(ts, segments, method_map)
    final_fc.to_csv('forecasts_next_{}_months.csv'.format(HORIZON))
    print("Forecasts and comparison files generated.")
```

```python
DATA_PATH = 'sales_data 2.csv'
HORIZON = 6  # months ahead
INITIAL_TRAIN_MONTHS = 24
ALPHA_CROSTON = 0.1
SEGMENT_THRESHOLDS = {'A': 0.8, 'B': 0.95, 'C': 1.0}
```

```python
import pandas as pd
import numpy as np
from statsmodels.tsa.holtwinters import ExponentialSmoothing
import warnings

warnings.filterwarnings('ignore')

# --- Configuration ---


# --- Data Loading and Preprocessing ---
def load_and_prepare(path):
    df = pd.read_csv(path, parse_dates=['Date'])
    df.set_index('Date', inplace=True)
    monthly = (
        df.groupby([pd.Grouper(freq='M'), 'Product'])['Revenue']
          .sum()
          .reset_index(name='Monthly_Revenue')
    )
    ts = monthly.pivot(index='Date', columns='Product', values='Monthly_Revenue').fillna(0)
    return ts

# --- ABC Segmentation ---
def abc_segmentation(ts):
    total_rev = ts.sum()
    cum_share = total_rev.sort_values(ascending=False).cumsum() / total_rev.sum()
    segments = {}
    for product, share in cum_share.items():
        if share <= SEGMENT_THRESHOLDS['A']:
            segments[product] = 'A'
        elif share <= SEGMENT_THRESHOLDS['B']:
            segments[product] = 'B'
        else:
            segments[product] = 'C'
    return segments

# --- Forecasting Methods ---
def fit_ets(series, horizon):
    series = series.asfreq('M')
    model = ExponentialSmoothing(series,
                                 trend='add',
                                 damped_trend=True,
                                 seasonal='add',
                                 seasonal_periods=12)
    fit = model.fit(optimized=True)
    return fit.fittedvalues, fit.forecast(horizon)


def croston_forecast(series, alpha, horizon):
    demand = series.values
    intervals = (demand > 0).astype(int)
    q = demand[demand > 0].mean() if demand[demand > 0].size > 0 else 0
    p = intervals.mean() if intervals.size > 0 else 0
    for t in range(1, len(demand)):
        if demand[t] > 0:
            q = alpha * demand[t] + (1 - alpha) * q
            p = alpha * intervals[t] + (1 - alpha) * p
    f_val = q / p if p > 0 else 0
    idx = [series.index[-1] + pd.DateOffset(months=i) for i in range(1, horizon+1)]
    return pd.Series([f_val] * horizon, index=idx)


def seasonal_naive_forecast(series, horizon):
    series = series.asfreq('M')
    if len(series) < 12:
        vals = np.tile(series.values, horizon)[:horizon]
    else:
        last_year = series[-12:]
        reps = (horizon // 12) + 1
        vals = np.tile(last_year.values, reps)[:horizon]
    idx = [series.index[-1] + pd.DateOffset(months=i) for i in range(1, horizon+1)]
    return pd.Series(vals, index=idx)

# --- Rolling Cross-Validation ---
def safe_mape(y_true, y_pred):
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    mask = y_true != 0
    if not mask.any():
        return np.nan
    return np.mean(np.abs((y_true[mask] - y_pred[mask]) / y_true[mask])) * 100


def rolling_cv(series, model_func, horizon=HORIZON, initial_train=INITIAL_TRAIN_MONTHS, alpha=ALPHA_CROSTON):
    n = len(series)
    folds = (n - initial_train) // horizon
    errors = []
    for i in range(folds):
        end = initial_train + i * horizon
        train = series[:end]
        test = series[end:end+horizon]
        # Generate forecast based on function signature
        if model_func == croston_forecast:
            forecast = model_func(train, alpha, horizon)
        else:
            forecast = model_func(train, horizon)
        y_true = test.values[:len(forecast)]
        y_pred = forecast.values[:len(test)]
        mape = safe_mape(y_true, y_pred)
        if not np.isnan(mape):
            errors.append(mape)
    return np.nanmean(errors)

# --- Model Selection for B-segment ---
def compare_methods_b(ts, segments, sample_n=5):
    results = []
    b_skus = [p for p, s in segments.items() if s == 'B'][:sample_n]
    for sku in b_skus:
        series = ts[sku]
        if len(series) < INITIAL_TRAIN_MONTHS + HORIZON:
            continue
        m_croston = rolling_cv(series, croston_forecast)
        m_seasonal = rolling_cv(series, seasonal_naive_forecast)
        pref = 'Croston' if (not np.isnan(m_croston) and m_croston <= m_seasonal) else 'SeasonalNaive'
        results.append({
            'Product': sku,
            'Croston_MAPE': round(m_croston, 2),
            'SeasonalNaive_MAPE': round(m_seasonal, 2),
            'Preferred': pref
        })
    return pd.DataFrame(results)

# --- Final Forecasting Pipeline ---
def generate_forecasts(ts, segments, method_map=None, horizon=HORIZON):
    forecasts = []
    for product, seg in segments.items():
        series = ts[product]
        if seg == 'A':
            _, fc = fit_ets(series, horizon)
        elif seg == 'B':
            pref = method_map.get(product) if method_map else None
            if pref == 'SeasonalNaive':
                fc = seasonal_naive_forecast(series, horizon)
            else:
                fc = croston_forecast(series, ALPHA_CROSTON, horizon)
        else:
            fc = seasonal_naive_forecast(series, horizon)
        # Collect forecast with product as column
        fc.name = product
        forecasts.append(fc)
    # Concatenate ensuring alignment
    return pd.concat(forecasts, axis=1)

# --- Main Execution ---
if __name__ == '__main__':
    ts = load_and_prepare(DATA_PATH)
    segments = abc_segmentation(ts)

    # Evaluate B-segment methods and build override map
    comp_df = compare_methods_b(ts, segments)
    comp_df.to_csv('b_segment_method_comparison.csv', index=False)
    method_map = comp_df.set_index('Product')['Preferred'].to_dict()

    # Generate final forecasts
    final_fc = generate_forecasts(ts, segments, method_map)
    final_fc.to_csv(f'forecasts_next_{HORIZON}_months.csv')
    print("Forecasts and comparison files generated.")
```

```python
import pandas as pd
import numpy as np
from statsmodels.tsa.holtwinters import ExponentialSmoothing
import warnings

warnings.filterwarnings('ignore')

# --- Configuration ---
DATA_PATH = 'sales_data 2.csv'
HORIZON = 6  # months ahead
INITIAL_TRAIN_MONTHS = 24
ALPHA_CROSTON = 0.1
SEGMENT_THRESHOLDS = {'A': 0.8, 'B': 0.95, 'C': 1.0}

# --- Data Loading and Preprocessing ---
def load_and_prepare(path):
    df = pd.read_csv(path, parse_dates=['Date'])
    df.set_index('Date', inplace=True)
    monthly = (
        df.groupby([pd.Grouper(freq='M'), 'Product'])['Revenue']
          .sum()
          .reset_index(name='Monthly_Revenue')
    )
    ts = monthly.pivot(index='Date', columns='Product', values='Monthly_Revenue').fillna(0)
    return ts

# --- ABC Segmentation ---
def abc_segmentation(ts):
    total_rev = ts.sum()
    cum_share = total_rev.sort_values(ascending=False).cumsum() / total_rev.sum()
    segments = {}
    for product, share in cum_share.items():
        if share <= SEGMENT_THRESHOLDS['A']:
            segments[product] = 'A'
        elif share <= SEGMENT_THRESHOLDS['B']:
            segments[product] = 'B'
        else:
            segments[product] = 'C'
    return segments

# --- Forecasting Methods ---
from prophet import Prophet  # Facebook Prophet for flexible seasonality and changepoints

def fit_prophet(series, horizon):
    """
    Fit a Prophet model to monthly series and forecast next `horizon` months.
    """
    df = series.asfreq('M').reset_index().rename(columns={'Date':'ds', series.name:'y'})
    m = Prophet(yearly_seasonality=True, weekly_seasonality=False, daily_seasonality=False)
    m.fit(df)
    future = m.make_future_dataframe(periods=horizon, freq='M')
    fc = m.predict(future)
    out = fc[['ds','yhat']].set_index('ds').yhat[-horizon:]
    out.index = pd.DatetimeIndex(out.index)
    return out

# --- Existing methods ---
def fit_ets(series, horizon):
    series = series.asfreq('M')
    model = ExponentialSmoothing(series,
                                 trend='add',
                                 damped_trend=True,
                                 seasonal='add',
                                 seasonal_periods=12)
    fit = model.fit(optimized=True)
    return fit.fittedvalues, fit.forecast(horizon)


def croston_forecast(series, alpha, horizon):
    demand = series.values
    intervals = (demand > 0).astype(int)
    q = demand[demand > 0].mean() if demand[demand > 0].size > 0 else 0
    p = intervals.mean() if intervals.size > 0 else 0
    for t in range(1, len(demand)):
        if demand[t] > 0:
            q = alpha * demand[t] + (1 - alpha) * q
            p = alpha * intervals[t] + (1 - alpha) * p
    f_val = q / p if p > 0 else 0
    idx = [series.index[-1] + pd.DateOffset(months=i) for i in range(1, horizon+1)]
    return pd.Series([f_val] * horizon, index=idx)


def seasonal_naive_forecast(series, horizon):
    series = series.asfreq('M')
    if len(series) < 12:
        vals = np.tile(series.values, horizon)[:horizon]
    else:
        last_year = series[-12:]
        reps = (horizon // 12) + 1
        vals = np.tile(last_year.values, reps)[:horizon]
    idx = [series.index[-1] + pd.DateOffset(months=i) for i in range(1, horizon+1)]
    return pd.Series(vals, index=idx)

# --- Rolling Cross-Validation ---
def safe_mape(y_true, y_pred):
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    mask = y_true != 0
    if not mask.any():
        return np.nan
    return np.mean(np.abs((y_true[mask] - y_pred[mask]) / y_true[mask])) * 100


def rolling_cv(series, model_func, horizon=HORIZON, initial_train=INITIAL_TRAIN_MONTHS, alpha=ALPHA_CROSTON):
    n = len(series)
    folds = (n - initial_train) // horizon
    errors = []
    for i in range(folds):
        end = initial_train + i * horizon
        train = series[:end]
        test = series[end:end+horizon]
        # Generate forecast based on function signature
        if model_func == croston_forecast:
            forecast = model_func(train, alpha, horizon)
        else:
            forecast = model_func(train, horizon)
        y_true = test.values[:len(forecast)]
        y_pred = forecast.values[:len(test)]
        mape = safe_mape(y_true, y_pred)
        if not np.isnan(mape):
            errors.append(mape)
    return np.nanmean(errors)

# --- Model Selection for A-segment ---
def compare_methods_a(ts, segments, sample_n=5, initial_train=INITIAL_TRAIN_MONTHS, horizon=HORIZON):
    results = []
    a_skus = [p for p, s in segments.items() if s == 'A'][:sample_n]
    for sku in a_skus:
        series = ts[sku]
        if len(series) < initial_train + horizon:
            continue
        # ETS
        ets_cv = rolling_cv(series, fit_ets, horizon=horizon, initial_train=initial_train)
        # Prophet
        # wrap fit_prophet to match signature: accepts (series,horizon)
        prophet_cv = rolling_cv(series, lambda x, h: fit_prophet(x, h), horizon=horizon, initial_train=initial_train)
        pref = 'ETS' if (not np.isnan(ets_cv) and ets_cv <= prophet_cv) else 'Prophet'
        results.append({
            'Product': sku,
            'ETS_MAPE': round(ets_cv,2),
            'Prophet_MAPE': round(prophet_cv,2),
            'Preferred': pref
        })
    return pd.DataFrame(results)

# --- Visualization Helper ---
import matplotlib.pyplot as plt

def visualize_forecast(series, horizon, methods, labels):
    """
    Plot actual series and forecasts from multiple methods.
    `methods` is list of callables that take (series,horizon) and return forecast Series.
    `labels` is list of same length.
    """
    plt.figure(figsize=(10,5))
    series.asfreq('M').plot(label='Actual')
    last_date = series.index[-1]
    for func, lab in zip(methods, labels):
        fc = func(series, horizon)
        plt.plot(fc.index, fc.values, marker='o', label=lab)
    plt.axvline(last_date, color='k', linestyle='--', label='Forecast Start')
    plt.legend()
    plt.title(f'Forecast Comparison for {series.name}')
    plt.show()

# --- Final Forecasting Pipeline ---
def generate_forecasts(ts, segments, method_map=None, horizon=HORIZON):
    forecasts = []
    for product, seg in segments.items():
        series = ts[product]
        if seg == 'A':
            _, fc = fit_ets(series, horizon)
        elif seg == 'B':
            pref = method_map.get(product) if method_map else None
            if pref == 'SeasonalNaive':
                fc = seasonal_naive_forecast(series, horizon)
            else:
                fc = croston_forecast(series, ALPHA_CROSTON, horizon)
        else:
            fc = seasonal_naive_forecast(series, horizon)
        # Collect forecast with product as column
        fc.name = product
        forecasts.append(fc)
    # Concatenate ensuring alignment
    return pd.concat(forecasts, axis=1)

# --- Main Execution ---
if __name__ == '__main__':
    ts = load_and_prepare("sales_data 2.csv")
    segments = abc_segmentation(ts)

    # Evaluate B-segment methods and build override map
    comp_b = compare_methods_b(ts, segments)
    comp_b.to_csv('b_segment_method_comparison.csv', index=False)
    method_map = comp_b.set_index('Product')['Preferred'].to_dict()

    # Evaluate A-segment methods
    comp_a = compare_methods_a(ts, segments)
    comp_a.to_csv('a_segment_method_comparison.csv', index=False)

    # Generate final forecasts
    final_fc = generate_forecasts(ts, segments, method_map)
    final_fc.to_csv(f'forecasts_next_{HORIZON}_months.csv')
    print("Forecasts and comparison files generated.")

    # Visualize forecasts for a sample A-segment SKU
    if not comp_a.empty:
        sample_sku = comp_a.loc[0, 'Product']
        print(f"Visualizing forecast comparison for sample A-SKU: {sample_sku}")
        methods = [lambda s, h: fit_ets(s, h)[1], fit_prophet]
        labels = ['ETS', 'Prophet']
        visualize_forecast(ts[sample_sku], HORIZON, methods, labels)
```

```python
import pandas as pd
import numpy as np
from statsmodels.tsa.holtwinters import ExponentialSmoothing
import warnings

warnings.filterwarnings('ignore')

# --- Configuration ---
DATA_PATH = 'sales_data 2.csv'
HORIZON = 6  # months ahead
INITIAL_TRAIN_MONTHS = 24
ALPHA_CROSTON = 0.1
SEGMENT_THRESHOLDS = {'A': 0.8, 'B': 0.95, 'C': 1.0}

# --- Data Loading and Preprocessing ---
def load_and_prepare(path):
    df = pd.read_csv(path, parse_dates=['Date'])
    df.set_index('Date', inplace=True)
    monthly = (
        df.groupby([pd.Grouper(freq='M'), 'Product'])['Revenue']
          .sum()
          .reset_index(name='Monthly_Revenue')
    )
    ts = monthly.pivot(index='Date', columns='Product', values='Monthly_Revenue').fillna(0)
    return ts

# --- ABC Segmentation ---
def abc_segmentation(ts):
    total_rev = ts.sum()
    cum_share = total_rev.sort_values(ascending=False).cumsum() / total_rev.sum()
    segments = {}
    for product, share in cum_share.items():
        if share <= SEGMENT_THRESHOLDS['A']:
            segments[product] = 'A'
        elif share <= SEGMENT_THRESHOLDS['B']:
            segments[product] = 'B'
        else:
            segments[product] = 'C'
    return segments

# --- Forecasting Methods ---
from prophet import Prophet  # Facebook Prophet for flexible seasonality and changepoints

def fit_prophet(series, horizon):
    """
    Fit a Prophet model to monthly series and forecast next `horizon` months.
    """
    df = series.asfreq('M').reset_index().rename(columns={'Date':'ds', series.name:'y'})
    m = Prophet(yearly_seasonality=True, weekly_seasonality=False, daily_seasonality=False)
    m.fit(df)
    future = m.make_future_dataframe(periods=horizon, freq='M')
    fc = m.predict(future)
    out = fc[['ds','yhat']].set_index('ds').yhat[-horizon:]
    out.index = pd.DatetimeIndex(out.index)
    return out

# --- Existing methods ---
def fit_ets(series, horizon):
    series = series.asfreq('M')
    model = ExponentialSmoothing(series,
                                 trend='add',
                                 damped_trend=True,
                                 seasonal='add',
                                 seasonal_periods=12)
    fit = model.fit(optimized=True)
    return fit.fittedvalues, fit.forecast(horizon)


def croston_forecast(series, alpha, horizon):
    demand = series.values
    intervals = (demand > 0).astype(int)
    q = demand[demand > 0].mean() if demand[demand > 0].size > 0 else 0
    p = intervals.mean() if intervals.size > 0 else 0
    for t in range(1, len(demand)):
        if demand[t] > 0:
            q = alpha * demand[t] + (1 - alpha) * q
            p = alpha * intervals[t] + (1 - alpha) * p
    f_val = q / p if p > 0 else 0
    idx = [series.index[-1] + pd.DateOffset(months=i) for i in range(1, horizon+1)]
    return pd.Series([f_val] * horizon, index=idx)


def seasonal_naive_forecast(series, horizon):
    series = series.asfreq('M')
    if len(series) < 12:
        vals = np.tile(series.values, horizon)[:horizon]
    else:
        last_year = series[-12:]
        reps = (horizon // 12) + 1
        vals = np.tile(last_year.values, reps)[:horizon]
    idx = [series.index[-1] + pd.DateOffset(months=i) for i in range(1, horizon+1)]
    return pd.Series(vals, index=idx)

# --- Rolling Cross-Validation ---
def safe_mape(y_true, y_pred):
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    mask = y_true != 0
    if not mask.any():
        return np.nan
    return np.mean(np.abs((y_true[mask] - y_pred[mask]) / y_true[mask])) * 100


def rolling_cv(series, model_func, horizon=HORIZON, initial_train=INITIAL_TRAIN_MONTHS, alpha=ALPHA_CROSTON):
    n = len(series)
    folds = (n - initial_train) // horizon
    errors = []
    for i in range(folds):
        end = initial_train + i * horizon
        train = series[:end]
        test = series[end:end+horizon]
        # Generate forecast based on function signature
        if model_func == fit_ets:
            # fit_ets returns (fitted, forecast)
            forecast = model_func(train, horizon)[1]
        elif model_func == croston_forecast:
            forecast = model_func(train, alpha, horizon)
        else:
            forecast = model_func(train, horizon)
        y_true = test.values[:len(forecast)]
        y_pred = forecast.values[:len(test)]
        mape = safe_mape(y_true, y_pred)
        if not np.isnan(mape):
            errors.append(mape)
    return np.nanmean(errors)

# --- Model Selection for B-segment ---
def compare_methods_b(ts, segments, sample_n=5):
    """
    Compare Croston vs. Seasonal Naive for B-segment SKUs via rolling CV.
    """
    results = []
    b_skus = [p for p, s in segments.items() if s == 'B'][:sample_n]
    for sku in b_skus:
        series = ts[sku]
        if len(series) < INITIAL_TRAIN_MONTHS + HORIZON:
            continue
        m_croston = rolling_cv(series, croston_forecast)
        m_seasonal = rolling_cv(series, seasonal_naive_forecast)
        pref = 'Croston' if (not np.isnan(m_croston) and m_croston <= m_seasonal) else 'SeasonalNaive'
        results.append({'Product': sku,
                        'Croston_MAPE': round(m_croston,2),
                        'SeasonalNaive_MAPE': round(m_seasonal,2),
                        'Preferred': pref})
    return pd.DataFrame(results)

# --- Model Selection for A-segment ---
def compare_methods_a(ts, segments, sample_n=5, initial_train=INITIAL_TRAIN_MONTHS, horizon=HORIZON):
    results = []
    a_skus = [p for p, s in segments.items() if s == 'A'][:sample_n]
    for sku in a_skus:
        series = ts[sku]
        if len(series) < initial_train + horizon:
            continue
        # ETS
        ets_cv = rolling_cv(series, fit_ets, horizon=horizon, initial_train=initial_train)
        # Prophet
        # wrap fit_prophet to match signature: accepts (series,horizon)
        prophet_cv = rolling_cv(series, lambda x, h: fit_prophet(x, h), horizon=horizon, initial_train=initial_train)
        pref = 'ETS' if (not np.isnan(ets_cv) and ets_cv <= prophet_cv) else 'Prophet'
        results.append({
            'Product': sku,
            'ETS_MAPE': round(ets_cv,2),
            'Prophet_MAPE': round(prophet_cv,2),
            'Preferred': pref
        })
    return pd.DataFrame(results)

# --- Visualization Helper ---
import matplotlib.pyplot as plt

def visualize_forecast(series, horizon, methods, labels):
    """
    Plot train/test actuals and forecasts from multiple methods.
    Ensures each forecast uses a distinct color and handles NaNs gracefully.
    """
    import itertools
    # Color cycle distinct from default to avoid overlap with train/test colors
    forecast_colors = itertools.cycle(['tab:red', 'tab:green', 'tab:orange', 'tab:purple'])

    s = series.asfreq('M')
    train = s[:-horizon]
    test = s[-horizon:]

    plt.figure(figsize=(10,5))
    plt.plot(train.index, train.values, label='Train', color='tab:blue')
    plt.plot(test.index, test.values, label='Test', color='black', linestyle='--')
    last_date = train.index[-1]
    plt.axvline(last_date, color='grey', linestyle=':')

    for func, lab in zip(methods, labels):
        try:
            fc = func(train, horizon)
        except Exception as e:
            print(f"Warning: {lab} forecast failed for {series.name}: {e}")
            continue
        # Drop NaNs to avoid invisible lines
        fc = fc.dropna()
        if fc.empty:
            print(f"Warning: {lab} forecast returned all NaNs for {series.name}.")
            continue
        plt.plot(fc.index, fc.values, marker='o', label=lab, color=next(forecast_colors), linewidth=2, zorder=5)
    plt.legend()
    plt.title(f'Forecast Comparison for {series.name}')
    plt.xlabel('Date')
    plt.ylabel('Revenue')
    plt.tight_layout()
    plt.show()

# --- Final Forecasting Pipeline ---
def generate_forecasts(ts, segments, method_map=None, horizon=HORIZON):
    forecasts = []
    for product, seg in segments.items():
        series = ts[product]
        if seg == 'A':
            _, fc = fit_ets(series, horizon)
        elif seg == 'B':
            pref = method_map.get(product) if method_map else None
            if pref == 'SeasonalNaive':
                fc = seasonal_naive_forecast(series, horizon)
            else:
                fc = croston_forecast(series, ALPHA_CROSTON, horizon)
        else:
            fc = seasonal_naive_forecast(series, horizon)
        # Collect forecast with product as column
        fc.name = product
        forecasts.append(fc)
    # Concatenate ensuring alignment
    return pd.concat(forecasts, axis=1)

# --- Main Execution ---
if __name__ == '__main__':
    # Load data and segment
    ts = load_and_prepare(DATA_PATH)
    segments = abc_segmentation(ts)

    # Evaluate B-segment methods and build override map
    comp_b = compare_methods_b(ts, segments)
    comp_b.to_csv('b_segment_method_comparison.csv', index=False)
    method_map = comp_b.set_index('Product')['Preferred'].to_dict()

    # Evaluate A-segment methods
    comp_a = compare_methods_a(ts, segments)
    comp_a.to_csv('a_segment_method_comparison.csv', index=False)

    # Generate final forecasts
    final_fc = generate_forecasts(ts, segments, method_map)
    final_fc.to_csv(f'forecasts_next_{HORIZON}_months.csv')
    print("Forecasts and comparison files generated.")

    # Visualize forecasts for a sample A-segment SKU
    if not comp_a.empty:
        sample_a = comp_a.loc[0, 'Product']
        print(f"A-segment comparison for: {sample_a}")
        methods_a = [lambda s, h: fit_ets(s, h)[1], fit_prophet]
        labels_a = ['ETS', 'Prophet']
        visualize_forecast(ts[sample_a], HORIZON, methods_a, labels_a)

    # Visualize forecasts for a sample B-segment SKU
    if not comp_b.empty:
        sample_b = comp_b.loc[0, 'Product']
        print(f"B-segment comparison for: {sample_b}")
        methods_b = [lambda s, h: croston_forecast(s, ALPHA_CROSTON, h), seasonal_naive_forecast]
        labels_b = ['Croston', 'SeasonalNaive']
        visualize_forecast(ts[sample_b], HORIZON, methods_b, labels_b)
```
